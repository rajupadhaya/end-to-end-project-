```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

# 1. Setup Data (Only your 4 features)
features = ["Open_Price", "High_Price", "Low_Price", "Volume"]
X_raw = df_final[features].copy()
y_raw = df_final["Asset_Type"].copy()

# 2. Transform to % Change (Crucial for realistic 0.80 accuracy)
X = X_raw.pct_change().replace([np.inf, -np.inf], np.nan).dropna()
y = y_raw.iloc[X.index] # Keep y aligned with X after dropping NaNs

# 3. Encode Target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 4. Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# 5. Train Model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model.fit(X_train, y_train)

# 6. Results
acc = model.score(X_test, y_test)
print(f"Realistic Accuracy: {acc:.4f}")

# -----------------------------
# 7. QUICK TESTER (Single Input)
# -----------------------------
def predict_now(open_p, high_p, low_p, vol):
    # This simulates a small % change for a single prediction
    test_data = np.array([[(high_p-open_p)/open_p, (high_p-low_p)/low_p, (low_p-open_p)/open_p, 0.02]])
    pred = model.predict(test_data)[0]
    return le.inverse_transform([pred])[0]

# Example test
print("Prediction Test:", predict_now(50000, 51000, 49500, 100000))
```
