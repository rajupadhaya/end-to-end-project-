```python
df12["Volatility"] = df12.groupby("Asset_Name")["Daily_Return"]\
                    .rolling(7).std().reset_index(level=0, drop=True)

# Target: Next Day Return
df12["Target_Return"] = df12.groupby("Asset_Name")["Daily_Return"].shift(-1)
```

```python
# ==========================================
# AI Investment Model (Improved Version)
# ==========================================

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor

# ==========================================
# 1. LOAD DATA
# ==========================================

df = df_final.copy()   # your cleaned dataset

# Fix duplicate date column if exists
if "date" in df.columns:
    df.drop(columns=["date"], inplace=True)

# Convert Date
df["Date"] = pd.to_datetime(df["Date"])

# Sort properly (VERY IMPORTANT)
df.sort_values(by=["Asset_Name", "Date"], inplace=True)

# ==========================================
# 2. FEATURE ENGINEERING (ADVANCED)
# ==========================================

# Moving Averages
df["MA_7"] = df.groupby("Asset_Name")["Close_Price"].rolling(7).mean().reset_index(0, drop=True)
df["MA_21"] = df.groupby("Asset_Name")["Close_Price"].rolling(21).mean().reset_index(0, drop=True)

# Price Range (intraday volatility)
df["Price_Range"] = df["High_Price"] - df["Low_Price"]

# Momentum
df["Momentum_5"] = df["Close_Price"] - df.groupby("Asset_Name")["Close_Price"].shift(5)

# Lag Features
df["Lag_1"] = df.groupby("Asset_Name")["Daily_Return"].shift(1)
df["Lag_2"] = df.groupby("Asset_Name")["Daily_Return"].shift(2)
df["Lag_3"] = df.groupby("Asset_Name")["Daily_Return"].shift(3)

# Rolling Returns
df["Return_7"] = df.groupby("Asset_Name")["Close_Price"].pct_change(7)
df["Return_21"] = df.groupby("Asset_Name")["Close_Price"].pct_change(21)

# Sharpe Ratio
df["Sharpe_Ratio"] = df["Daily_Return"] / (df["Volatility"] + 1e-6)

# TARGET: Future Return (Next Day)
df["Target"] = df.groupby("Asset_Name")["Daily_Return"].shift(-1)

# ==========================================
# 3. CLEAN DATA
# ==========================================

df.replace([np.inf, -np.inf], 0, inplace=True)
df.fillna(0, inplace=True)

# ==========================================
# 4. ENCODE CATEGORICAL DATA
# ==========================================

le_asset = LabelEncoder()
le_type = LabelEncoder()
le_sector = LabelEncoder()

df["Asset_Name_Enc"] = le_asset.fit_transform(df["Asset_Name"])
df["Asset_Type_Enc"] = le_type.fit_transform(df["Asset_Type"])
df["Sector_Enc"] = le_sector.fit_transform(df["Sector"])

# ==========================================
# 5. SELECT FEATURES
# ==========================================

features = [
    "Open_Price", "High_Price", "Low_Price", "Close_Price",
    "Volume", "Daily_Return", "Volatility",
    "MA_7", "MA_21",
    "Price_Range", "Momentum_5",
    "Lag_1", "Lag_2", "Lag_3",
    "Return_7", "Return_21",
    "Sharpe_Ratio",
    "Asset_Name_Enc", "Asset_Type_Enc", "Sector_Enc"
]

X = df[features]
y = df["Target"]

# ==========================================
# 6. TRAIN-TEST SPLIT (Time-Aware)
# ==========================================

split_index = int(len(df) * 0.8)

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# ==========================================
# 7. MODEL (XGBoost Advanced)
# ==========================================

model = XGBRegressor(
    n_estimators=400,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8
)

model.fit(X_train, y_train)

# ==========================================
# 8. EVALUATION
# ==========================================

y_pred = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R2 Score:", r2)

# ==========================================
# 9. PREDICTIONS
# ==========================================

df["Predicted_Return"] = model.predict(X)

# ==========================================
# 10. BUILD RECOMMENDATION ENGINE
# ==========================================

asset_summary = df.groupby("Asset_Name").agg({
    "Predicted_Return": "mean",
    "Volatility": "mean"
}).reset_index()

# Risk classification
def risk_level(vol):
    if vol < 0.01:
        return "Low"
    elif vol < 0.03:
        return "Medium"
    else:
        return "High"

asset_summary["Risk_Level"] = asset_summary["Volatility"].apply(risk_level)

# ==========================================
# 11. RECOMMEND FUNCTION
# ==========================================

def recommend(amount, risk):

    filtered = asset_summary[asset_summary["Risk_Level"] == risk]

    filtered = filtered.sort_values(by="Predicted_Return", ascending=False)

    total = filtered["Predicted_Return"].sum()

    filtered["Weight"] = filtered["Predicted_Return"] / total
    filtered["Investment"] = filtered["Weight"] * amount

    return filtered[["Asset_Name", "Investment", "Predicted_Return"]]

# ==========================================
# 12. TEST
# ==========================================

result = recommend(200, "Medium")

print("\nRecommended Portfolio:\n")
print(result)
```
